```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('winequalityN.csv')
df
```

```python
print("Dataset information:")
df.info()
```

```python
print("Missing values per column:")
df.isnull().sum()
```

```python
print("Descriptive statistics:")
df.describe()
```

```python
duplicates = df.duplicated().sum()
print(f"\nNumber of duplicated rows: {duplicates}")
if duplicates > 0:
    df = df.drop_duplicates()
    print("Duplicates removed successfully.")
else:
    print("No duplicates found.")
```

```python
print("\nMissing values before imputation:")
print(df.isna().sum())
```

```python
df = df.fillna(df.median(numeric_only=True))
```

```python
print("\nMissing values after imputation:")
print(df.isna().sum())
```

```python
if (df.select_dtypes(include='number') < 0).any().any():
    print("Warning: Some negative values found.")
else:
    print("No negative numeric values found.")

print("\nData cleaning completed. The dataset is ready for analysis.")
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["alcohol"], bins=20, color='lightblue', edgecolor='black')
plt.title("Distribution of Alcohol")
plt.xlabel("Alcohol (%)")
plt.ylabel("Frequency")
plt.grid(axis='y', alpha=0.3)
plt.show()
```
